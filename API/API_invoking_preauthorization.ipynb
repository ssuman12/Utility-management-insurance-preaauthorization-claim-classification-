```python
import requests
def make_api_request(data):
    # url for api
    url = 'http://127.0.0.1:10001/make_prediction'
    # make post request
    r = requests.post(url,data)
#     return r.text
    return r.json()
```

```python
rr = make_api_request(test_df.to_json())
df_rr = pd.read_json(json.dumps(rr))
df_rr.head() 
```

```python
def Fraud_label(answer):
    if answer=='Yes':
        return 1
    else:
        return 0
```

```python
df_rr['actual_label']=df_rr.Fraud.map(lambda x: Fraud_label(x))
```
